```python
import os
import random
base_path = "/samba/HDD/dataset_elec_oneplustwo/"
class_list = os.listdir(base_path)
for i in range(len(class_list)):
    class_path = base_path + class_list[i]
    subject_list = os.listdir(class_path)
    old_list = subject_list[:]
    #print(old_list)
    random.shuffle(subject_list)
    #print(subject_list)
    for j in range(len(subject_list)):
        old_path = class_path + "/" + old_list[j]
        new_path = class_path + "/" +"subject_" +  subject_list[j]
        print("*************************\n",old_path,"\n",new_path)
       #os.rename(old_path,new_path)
    
```
